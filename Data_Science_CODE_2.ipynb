```python
#making a correlation matrix between the different airlines to find smth
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# URLs for the datasets
url1 = "https://web.mit.edu/airlinedata/www/2020%2012%20Month%20Documents/Revenue%20and%20Related/Total%20Revenue/System%20Total%20Operating%20Revenue.htm" #Operating Revenue
url2 = "https://web.mit.edu/airlinedata/www/2020%2012%20Month%20Documents/Revenue%20and%20Related/Passenger%20Revenue/System%20Passenger%20Revenue.htm" #Passenger Revenue
url3 = "https://web.mit.edu/airlinedata/www/2020%2012%20Month%20Documents/Revenue%20and%20Related/Other%20Revenue/System%20Total%20Transport%20Related%20Revenue.htm" #Transport Revenue
url4 = "https://web.mit.edu/airlinedata/www/2020%2012%20Month%20Documents/Revenue%20and%20Related/Other%20Revenue/Ancillary%20Fee%20Revenue.htm"#Baggage Fee Revenue
url5 = "https://web.mit.edu/airlinedata/www/2020%2012%20Month%20Documents/Revenue%20and%20Related/Total%20Revenue/System%20Total%20Revenue%20(Ex-%20Transport%20Related)%20per%20ASM.htm" #Ex-Transport Revenue

table1 = pd.read_html(url1, encoding='latin1')
table2 = pd.read_html(url2, encoding='latin1')
table3 = pd.read_html(url3, encoding='latin1')
table4 = pd.read_html(url4, encoding='latin1')
table5 = pd.read_html(url5, encoding='latin1')


df1 = table1[0]
df2 = table2[0]
df3 = table3[0]
df4 = table4[0]
df5 = table5[0]

rows_to_remove = [0,1,2,4,13,21,26,27,29,31,32,33]

df1 = df1.drop(index=rows_to_remove)
df1 = df1.dropna(axis=1, how='all')
df1.reset_index(drop=True, inplace=True)

df1.columns = df1.iloc[0]
df1 = df1[1:]
df1.reset_index(drop=True, inplace=True)



df2 = df2.drop(index=rows_to_remove)
df2 = df2.dropna(axis=1, how='all')
df2.reset_index(drop=True, inplace=True)

df2.columns = df2.iloc[0]
df2 = df2[1:]
df2.reset_index(drop=True, inplace=True)


df3 = df3.drop(index=rows_to_remove)
df3 = df3.dropna(axis=1, how='all')
df3.reset_index(drop=True, inplace=True)

df3.columns = df3.iloc[0]
df3 = df3[1:]
df3.reset_index(drop=True, inplace=True)


new_rows_to_remove = [0,1,2,4,13,21,26,27,29,30,31,32,33]
df4 = df4.drop(index=new_rows_to_remove)
df4 = df4.dropna(axis=1, how='all')
df4.reset_index(drop=True, inplace=True)

df4.columns = df4.iloc[1]
df4 = df4[1:]
df4.reset_index(drop=True, inplace=True)


df5 = df5.drop(index=new_rows_to_remove)
df5 = df5.dropna(axis=1, how='all')
df5.reset_index(drop=True, inplace=True)
df5.columns = df5.iloc[0]
df5 = df5[1:]
df5.reset_index(drop=True, inplace=True)
```

```python
#linear regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

#printing the years
years = df1.columns[1:]
print(years)


x = df1_totalsector[1:].dropna().values.tolist()

#changing the list into an array
years = np.array(years).reshape(-1, 1)
x= np.array(x).reshape(-1, 1)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(years, x)

x_pred = np.array(years).reshape(-1, 1)
y_pred = model.predict(x_pred)

mse = mean_squared_error(x, y_pred)
r2 = r2_score(x, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
