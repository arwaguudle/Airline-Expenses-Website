```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#importing the table (which was found online)
url = "https://web.mit.edu/airlinedata/www/2020%2012%20Month%20Documents/Revenue%20and%20Related/Total%20Revenue/System%20Total%20Operating%20Revenue.htm"
# The encoding was being misinterpreted. Instead of 'latin-1', use 'latin1', or 'iso-8859-1'.
tables = pd.read_html(url, encoding='latin1')

# Assuming the first table is the one you want
#this data is from System Total Operating Expenses (i couldnt find it from the airline )
df = tables[0]
#want to remove some unwanted rows and columns

rows_to_remove = [0,1,2,4,13,20,26,27,29,31,32,33]

df = df.drop(index=rows_to_remove)
df = df.dropna(axis=1, how='all')
df.reset_index(drop=True, inplace=True)

df.columns = df.iloc[0]
df = df[1:]
df.reset_index(drop=True, inplace=True)

df
```

```python
#gonna try and make a multiple line graphs comparing the different subs

sub_network = df.iloc[7]
sub_hybrid = df.iloc[13]
sub_ulcc = df.iloc[18]
#sub_network = pd.to_numeric(sub_network, errors='coerce')

# Convert the index to string to avoid the error
#sub_network.index = sub_network.index.astype(str)
a = sub_network.index[1:]
years = np.array(a, dtype=float)
print(years)
```

```python
b =sub_network.values[1:]
sub_network = np.array([float(value.replace('$', '').strip()) for value in b])
print(sub_network)

c =sub_hybrid.values[1:]
sub_hybrid = np.array([float(value.replace('$', '').strip()) for value in c])
print(sub_hybrid)

d =sub_ulcc.values[1:]
sub_ulcc = np.array([float(value.replace('$', '').strip()) for value in d])
print(sub_ulcc)
```

```python
plt.figure(figsize=(10,6))
plt.plot(years,sub_network)
plt.plot(years,sub_hybrid)
plt.plot(years,sub_ulcc)
plt.xlabel("years")
plt.ylabel("System Total Operating Revenue (Billions)")
plt.legend(["sub network","sub hybrid","sub ulcc"], loc="upper left")
```

```python
#how to save the picture in my file
plt.figure(figsize=(10, 6))
plt.savefig("sub_network.png")
```

```python
#making a barchart of the total sectors in relation to the time
total_prices = df.iloc[20]
total_prices.index = total_prices.index.astype(str)
total_prices
```
