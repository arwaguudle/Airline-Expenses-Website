```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


#importing the table (which was found online)
url = "https://web.mit.edu/airlinedata/www/2019%2012%20Month%20Documents/Expense%20Related/Total/System%20Total%20Operating%20Expenses.htm"
tables = pd.read_html(url)

# Assuming the first table is the one you want
#this data is from System Total Operating Expenses (i couldnt find it from the airline )
df = tables[0]
#want to remove some unwanted rows and columns
rows_to_remove = [0,1,2,4,13,20,26,27,29,31,32,33,34]

df = df.drop(index=rows_to_remove)
df = df.dropna(axis=1, how='all')
df.reset_index(drop=True, inplace=True)

df.columns = df.iloc[0]
df = df[1:]
df.reset_index(drop=True, inplace=True)
df



```

```python
#gonna try and make a multiple line graphs comparing the different subs

sub_network = df.iloc[7]
sub_lcc = df.iloc[13]
sub_other = df.iloc[18]
#sub_network = pd.to_numeric(sub_network, errors='coerce')

# Convert the index to string to avoid the error
#sub_network.index = sub_network.index.astype(str)
a = sub_network.index[1:]
years = np.array(a, dtype=float)
print(years)
```

```python
b =sub_network.values[1:]
sub_network = np.array([float(value.replace('$', '').strip()) for value in b])
print(sub_network)

c =sub_lcc.values[1:]
sub_lcc = np.array([float(value.replace('$', '').strip()) for value in c])
print(sub_lcc)

d =sub_other.values[1:]
sub_other = np.array([float(value.replace('$', '').strip()) for value in d])
print(sub_other)
```

```python
plt.plot(years,sub_network)
plt.plot(years,sub_lcc)
plt.plot(years,sub_other)
plt.xlabel("years")
plt.ylabel("System Total Operating Expenses (Billions)")
plt.legend(["sub_network","sub_lcc","sub_other"], loc="upper left")
```

```python
#how to save the picture in my file
plt.figure(figsize=(10, 6))
plt.savefig("sub_network.png")
```

```python
#making a barchart of the total sectors in relation to the time
total_prices = df.iloc[20]
total_prices.index = total_prices.index.astype(str)
total_prices
```

```python
e =total_prices.values[1:]
total_prices = np.array([float(value.replace('$', '').strip()) for value in e])
print(total_prices)
```

```python
#making the bar chart
plt.bar(years,total_prices,color="pink")
plt.xlabel("years")
plt.ylabel("System Total Operating Expenses (Billions)")
plt.title("Total Sector Expenses")
plt.show()
```

```python
#looking at the specific years (2002)
#in column 6

two_two = df.iloc[:,8]
airline_names = df.iloc[:, 0]
combined_df = pd.DataFrame({
    "Airline Name": airline_names,
    "2002": two_two
})
combined_df


#excluding some rows
rows_to_remove = [7,13,18,19,20]
combined_df = combined_df.drop(index=rows_to_remove)
combined_df

# getting rid of the $
combined_df['2002'] = combined_df['2002'].str.replace('$', '')
combined_df['2002'] = combined_df['2002'].str.replace(',', '')
combined_df

#making the empty figures in row 12 and row 17 into 0
combined_df.iloc[11, 1] = 0
combined_df.iloc[15, 1] = 0

combined02 = combined_df
combined02
```

```python
#doing the same thing for the years where
#looking at the specific years (2010)

two_ten = df.iloc[:,16]
airline_names = df.iloc[:, 0]
combined_df = pd.DataFrame({
    "Airline Name": airline_names,
    "2010" : two_ten
})

#excluding some rows
rows_to_remove = [7,13,18,19,20]
combined_df = combined_df.drop(index=rows_to_remove)

# getting rid of the $
combined_df['2010'] = combined_df['2010'].str.replace('$', '')
combined_df['2010'] = combined_df['2010'].str.replace(',', '')

#making the empty figures in row 12 and row 17 into 0
combined_df.iloc[3,1] = 0
combined_df.iloc[6,1] = 0

combined10 = combined_df
combined10
```

```python
#doing the same thing for the years where
#looking at the specific years (2014)

fourteen = df.iloc[:,20]
airline_names = df.iloc[:, 0]
combined_df = pd.DataFrame({
    "Airline Name": airline_names,
    "2014" : fourteen
})

#excluding some rows
rows_to_remove = [7,13,18,19,20]
combined_df = combined_df.drop(index=rows_to_remove)

# getting rid of the $
combined_df['2014'] = combined_df['2014'].str.replace('$', '')
combined_df['2014'] = combined_df['2014'].str.replace(',', '')

#making the empty figures in row 12 and row 17 into 0
combined_df.iloc[1,1] = 0
combined_df.iloc[3,1] = 0
combined_df.iloc[6,1] = 0
combined_df.iloc[9,1] = 0

combined14 = combined_df
combined14
```
